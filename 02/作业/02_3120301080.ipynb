```python
def two():
    x = np.arange(0.0,2.0,0.1)
    y1 = np.sin(np.pi * x)
    y2 = x * 2
    plt.subplot(2,1,1)
    plt.plot(x,y1,'yo-')
    plt.title('two')
    plt.xlabel('x1')
    plt.ylabel('y1')
    plt.grid()

    plt.subplot(2,1,2)
    plt.plot(x,y2,'r.-')
    plt.xlabel('x2')
    plt.ylabel('y2')
    plt.grid()
    plt.show()
two() 
```

```python
def d():
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1,projection='3d')
    x = np.arange(1,10,1)
    y = np.arange(1,10,1)
    x,y = np.meshgrid(x,y)
    z = 3*x +2 *y +30
    surf=ax.plot_surface(x,y,z,rstride=1,cstride=1,linewidth=0,antialiased=True)
    ax.set_xlabel('x',color='r')
    ax.set_ylabel('y',color='g')
    ax.set_zlabel('z',color='b')
    ax.set_zlim3d(0,100)
    fig.colorbar(surf,shrink=0.5,aspect=5)
    plt.show()
d()
```

```python
url = 'https://timgsa.baidu.com/timg?image&quality=80&size=b9999_10000&sec=1608130287501&di=30541ace1882fcab100b4eadbc3d550b&imgtype=0&src=http%3A%2F%2Fpic3.16pic.com%2F00%2F01%2F11%2F16pic_111395_b.jpg'
root = 'E://a//'
path = root + url.split('/')[-1]
try:
    if not os.path.exists(root):
        os.mkdir(root)
    if not os.path.exists(path):
        r = requests.get(url)
        with open(path,'wb') as f:
            f.write(r.content)
            f.close()
            print('保存成功')
    else:
        print('图片已存在')
except:
    print('保存失败')
```
