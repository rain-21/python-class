```python
# 分项的权重


def W(N, D, i):
    d = 0
    for j in range(N):
        d += D[j]
    w = (1-D[i]/d)/(N - 1)

    return w

```

```python
# 标准欧式距离
import numpy as np


def D(Z, G_i, t_j, LenZ, LenG):
    d = 0
    Var = np.var(G_i)
    for k in range(LenZ):
        if (t_j + k) >= LenG:
            break
        else:
           d += pow(Z[k] - G_i[t_j + k], 2)

    d = d/Var

    return d

```
