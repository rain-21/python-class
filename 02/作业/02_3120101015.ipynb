```python
def my_function(x):
  return 5 * x        #使函数返回值
print(my_function(2))

#递归函数
def tri_recursion(k):
  if k>0:
    result = k+tri_recursion(k-1)
    print(result,end=" ")
  else:
    result = 0
  return result
print("\nRecursion Example Results")
tri_recursion(6)

#lambda函数
x = lambda a, b, c : a + b + c
print("\n",x(5, 6, 2))

def myfunc(n):
  return lambda a : a * n          #lambda函数能将参数传进去并参加运算
mytripler = myfunc(3)
print(mytripler(11))

#创建类和对象
class MyClass:
  x = 7
p1 = MyClass()
print(p1.x)
#__init__()函数
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
p1 = Person("Bill", 23)
print(p1.name)
print(p1.age)
#创建对象方法
class person():
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def function(self):
        print("My name is "+self.name,"and I am "+ self.age +" years old.")
p2 = p3 = person("bill","23")
p2.function()
del p3
# print(p3)    p3已被删除，此时找不到这个变量

#创建迭代器
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self
  def __next__(self):
    x = self.a
    self.a += 1
    return x

myclass = MyNumbers()
myiter = iter(myclass)
print(next(myiter))
print(next(myiter))

#停止迭代器
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)
for x in myiter:
    print(x,end=" ")
```
