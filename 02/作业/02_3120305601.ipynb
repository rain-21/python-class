```python
from PIL import Image
import numpy as np
def translation_tansform(img,dx,dy):
    h, w ,c = img.shape
    res = np.zeros((h,w,c))
    for x in range(h):
        for y in range(w):
            _x = x + dx
            _y = y + dy
            if _x < h and _y < w:
                res[_x][_y] = img[x][y]
    print(res)
    img = Image.fromarray(np.uint8(res)).convert('RGB')
    return img

def rotate_transform(img,angle):
    h,w,c = img.shape
    len = int(np.sqrt(h**2 + w**2 + 0.5))
    res = np.zeros((len,len,c))
    anglePI = angle * np.pi / 180
    top = int((h/2)*np.cos(anglePI)+(w/2)*np.sin(anglePI)+0.5)
    left =int((w/2)*np.cos(anglePI)-(1.5*h)*np.sin(anglePI)+0.5)
    for x in range(len):
        for y in range(len):
            _x = int((x+top)*np.cos(anglePI)-(left + y)*np.sin(anglePI)-h/2+0.5)
            _y = int((x+top)*np.sin(anglePI)+(left + y)*np.cos(anglePI)-w/2+0.5)
            if _x < h and _x>=0 and _y < w and _y >=0:
                res[x][y] = img[_x][_y]

    img = Image.fromarray(np.uint8(res)).convert('RGB')
    return img



img = Image.open('cat.jpeg').convert('RGB')

img = np.array(img)
# img = translation_tansform(img,50,50)

# img.show()
img = rotate_transform(img,60)
img.show()
```

```python
import numpy as np
from PIL import Image
from glob import glob
def bl_interpolate(img, ax, ay):
    h,w,c = img.shape
    print(img.shape)
    ah = int(ay*h)
    aw = int(ax*w)
    res = np.zeros((ah,aw,c))
    for x in range(ah):
        for y in range(aw):
            _x = int(min(np.floor(x/ax),h-2))
            _y = int(min(np.floor(y/ay),w-2))
            # print(_x,_y)
            dx = x/ax - _x
            dy = y/ay - _y

            res[x][y] = (1-dx)*(1-dy)*img[_x][_y]+(1-dx)*dy*img[_x][_y+1] + dx*(1-dy)*img[_x+1][_y]+dx*dy*img[_x+1][_y+1]
    res = np.clip(res,0,255)
    img = Image.fromarray(np.uint8(res)).convert('RGB')
    return img
img = Image.open('cat.jpeg').convert('RGB')
print(img.size)

img = np.array(img)
scale_factor = 0.5
for i in range(1,5):
    img = bl_interpolate(img,scale_factor,scale_factor)
    img.save('img_{}.jpg'.format(scale_factor**i))
    img = np.array(img)

img = Image.open('img_0.0625.jpg')
img = np.array(img)
img = bl_interpolate(img,2,2)
# img_list = glob('*.jpg')
# print(img_list)
# for img_name in img_list:
#     scale_factor = float(img_name.split('.jpg')[0].split('_')[1])
#     print(scale_factor)
#     img = Image.open(img_name).convert('RGB')
#     img = np.array(img)
#     img_recover = bl_interpolate(img,2,2)
#     img_recover = np.array(img_recover)
#     if img_name == 'img_0.5.jpg':
#         img_original = Image.open('cat.jpeg')
#     else:
#         img_original = Image.open('img_{}.jpg'.format(scale_factor*2))
#     img_original = np.array(img_original)
#     h = int(min(img_original.shape[0],img_recover.shape[0]))
#     w = int(min(img_original.shape[1],img_recover.shape[1]))
#     img_original = img_original[:h,:w]
#     img_recover = img_recover[:h,:w]
#     res = img_original - img_recover
#     res = np.clip(res,0,255)
#     res_img = Image.fromarray(np.uint8(res)).convert('RGB')
#     res_img.save('img_res{}.jpg'.format(scale_factor*2))
```

```python
img_list = glob('*.jpg')
print(img_list)
for img_name in img_list:
    scale_factor = float(img_name.split('.jpg')[0].split('_')[1])
    print(scale_factor)
    img = Image.open(img_name).convert('RGB')
    img = np.array(img)
    img_recover = bl_interpolate(img,2,2)
    img_recover = np.array(img_recover)
    if img_name == 'img_0.5.jpg':
        img_original = Image.open('cat.jpeg')
    else:
        img_original = Image.open('img_{}.jpg'.format(scale_factor*2))
    img_original = np.array(img_original)
    h = int(min(img_original.shape[0],img_recover.shape[0]))
    w = int(min(img_original.shape[1],img_recover.shape[1]))
    img_original = img_original[:h,:w]
    img_recover = img_recover[:h,:w]
    res = img_original - img_recover
    res = np.clip(res,0,255)
    res_img = Image.fromarray(np.uint8(res)).convert('RGB')
    res_img.save('img_res{}.jpg'.format(scale_factor*2))
```

```python
fruits = ['apples', 'oranges', 'pears', 'apricots']
change = [1, 'pennies', 2, 'dimes', 3, 'quarters']

# this first kind of for-loop goes through a list
for number in range(10):
    print(f"This is count {number}")

# same as above
for fruit in fruits:
    print(f"A fruit of type: {fruit}")

# also we can go through mixed lists too
# notice we have to use {} since we don't know what's in it
for i in change:
    print(f"I got {i}")

# we can also build lists, first start with an empty one
elements = []

# then use the range function to do 0 to 5 counts
for i in range(0, 6):
    print(f"Adding {i} to the list.")
    # append is a function that lists understand
    elements.append(i)

# now we can print them out too
for i in elements:
    print(f"Element was: {i}")
```
