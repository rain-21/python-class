```python
def two_numbers_operation(num1,num2):
    sum = num1 + num2
    diffenrence = num1 - num2
    product = num1 * num2
    division = num1 / num2
    exponetial = num1 ** num2
    remainder = num1 % num2
    print(f"两数之和为：{sum}")
    print(f"两数之差为：{diffenrence}")
    print(f"两数之积为：{product}")
    print(f"两数比值为：{division}")
    print(f"第一个数的幂：{exponetial}")
    print(f"第一个数除以第二个数的余数为：{remainder}")
two_numbers_operation(2, 3)
```

```python
input_file = 'sample.txt'
def print_all(f):
    print(f.read())
def rewind(f):
    f.seek(0)
def print_a_line(line_count, f):
    print(line_count, f.readline())
current_file = open(input_file)
print("First let's print the whole file:\n")
print_all(current_file)
print("Now let's rewind, kind of like a tape.")
rewind(current_file)

print("Let's print three lines:")
current_line = 1
while current_line < 4:
    print_a_line(current_line, current_file)
    current_line += 1

current_file.close()
```

```python
def add(a, b):
    print(f"ADDING {a} + {b}")
    return a + b
def subtract(a, b):
    print(f"SUBTRACTING {a} - {b}")
    return a - b
def multiply(a, b):
    print(f"MULTIPLYING {a} * {b}")
    return a * b
def divide(a, b):
    print(f"DIVIDING {a} / {b}")
    return a / b
print("Let's do some math with just functions!")
age = add(30, 5)
height = subtract(78, 4)
weight = multiply(90, 2)
iq = divide(100, 2)
print(f"Age: {age}, Height: {height}, Weight: {weight}, IQ: {iq}")
print("Here is a puzzle.")
what = add(age, subtract(height, multiply(weight, divide(iq, 2))))
print("That becmones:", what, "Can you do it by hand?")
```

```python
# 导入random用于后面生成随机数
import random
# 导入math用于后面sin和pi的计算
import math
# Matplotlib 是一个 Python 的 2D绘图
import matplotlib.pyplot as plt
# numpy 支持高级大量的维度数组与矩阵运算
import numpy as np

# 设定初始列表，初始化预测值，期望值等
X = []
Y = []
f1_yuce = []
f1_qiwang = 0
var1 = 0

f2_yuce = []
f2_qiwang = 0
var2 = 0

variances1 = []
biases1 = []
error1 = []
variances2 = []
biases2 = []
error2 = []

# 设定训练次数
xlcs = 10

# 计算两种模型对测试数据x所产生的预测值及期望值
for p in range(10):
    # 利用随机数生成测试数据(x,y)，并打印
    print('第', p, '次生成测试数据：')
    x = random.uniform(-1, 1)
    y = math.sin(math.pi * x)
    print('测试样本x,y:', x, y)

    for i in range(xlcs):
        # 生成训练数据(x1,y1),(x2,y2)
        for j in range(2):
            X.append(random.uniform(-1, 1))
            Y_temp = math.sin(math.pi*X[j])
            Y.append(Y_temp)

        # 计算第一个模型H0=b对测试数据x的预测输出
        f1_yuce.append((Y[0]+Y[1])/2)
        f1_qiwang = f1_qiwang+f1_yuce[i]

        # 计算第一个模型H0=b对测试数据x的预测输出
        a = (Y[1]-Y[0])/(X[1]-X[0])
        b = (Y[0]*X[1]-Y[1]*X[0])/(X[1]-X[0])
        f2_yuce.append(a*x+b)
        f2_qiwang = f2_qiwang + f2_yuce[i]
        # 删除生成的训练数据集，以便于后面存入新的一组训练数据，重复训练
        del X[1]
        del X[0]
        del Y[1]
        del Y[0]

    # 分别计算两种模型下预测值的期望，并打印
    f1_qiwang = f1_qiwang/xlcs
    f2_qiwang = f2_qiwang/xlcs
    print('两种模型对同一个x的预测值的期望为（模型H0为绿色，模型H1为蓝色）：', f1_qiwang, f2_qiwang)

    # 计算方差之和，并将每一次的预测值在坐标纸上做出散点图
    for k in range(xlcs):
        var1 = var1 + (f1_yuce[k] - f1_qiwang) ** 2
        var2 = var2 + (f2_yuce[k] - f2_qiwang) ** 2
        plt.scatter(x, f1_yuce[k], c='green', s=25, alpha=0.5, marker='4')
        plt.scatter(x, f2_yuce[k], c='blue', s=25, alpha=0.5, marker='*')

    # 分别计算两种模型下的方差，偏差，泛化误差，并打印
    var1 = var1/xlcs
    bias1 = (f1_qiwang-y)**2
    err1 = bias1+var1

    var2 = var2 / xlcs
    bias2 = (f2_qiwang - y) ** 2
    err2 = bias2 + var2

    print('第一个模型模型H0(x)=b的方差，偏差，泛化误差分别为：', var1, bias1, err1)
    print('第二个模型模型H1(x)=ax+b的方差，偏差，泛化误差分别为：', var2, bias2, err2)

    # 做sin的图像
    xdata = np.linspace(-1, 1)
    ydata = np.sin(xdata*math.pi)
    plt.plot(xdata, ydata, 'r', label='target_line')
    plt.legend()

    # 将期望值在坐标纸上描出
    plt.scatter(x, f1_qiwang, c='green', s=50, marker=',', label='H0')
    plt.scatter(x, f2_qiwang, c='blue', s=50, marker=',', label='H1')
    plt.legend()
    plt.scatter(x, y, c='red', s=10)

    # 设置图表标题，并给坐标轴加上标签
    plt.title('target function and predictions of two models', fontsize=18)
    plt.xlabel('x', fontsize=14)
    plt.ylabel('y', fontsize=14)
    plt.show()

    variances1.append(var1)
    biases1.append(bias1)
    error1.append(err1)
    variances2.append(var2)
    biases2.append(bias2)
    error2.append(err2)
# 将方差、偏差、泛化误差转化为数组，以便于画图
x = np.array(range(10))
y_var1 = np.array(variances1)
y_bia1 = np.array(biases1)
y_err1 = np.array(error1)
y_var2 = np.array(variances2)
y_bia2 = np.array(biases2)
y_err2 = np.array(error2)

#画出两种模型的偏差，方差，泛化误差折线图，以便于对比
plt.plot(x, y_var1, c='red', marker=',', alpha=0.2, label='HO-variance')
plt.plot(x, y_bia1, c='blue', marker='*', alpha=0.2, label='HO-bias')
plt.plot(x, y_err1, c='green', marker='8', alpha=0.2, label='HO-error')
plt.plot(x, y_var2, c='red', marker=',', alpha=0.8, label='H1-variance')
plt.plot(x, y_bia2, c='blue', marker='*', alpha=0.8, label='H1-bias')
plt.plot(x, y_err2, c='green', marker='8', alpha=0.8, label='H1-error')
plt.legend()
# 设置图表标题，并给坐标轴加上标签
plt.title('comparison of two models', fontsize=18)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.show()
```
