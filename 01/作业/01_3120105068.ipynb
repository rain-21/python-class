```python
# 随机生成两个列表，其中存储1~10的整数，将两个列表中元素之和为10的那两个元素及其位置输出
# 导入random模块以生成随机数
import random

#创建两个空列表A和B，以便于后面存储元素
A = []
B = []

"""
利用for循环给列表A和B中分别添加随机整数
并将两个列表打印出来
"""
for i in range(10):
    A.append(random.randint(1, 10))
    B.append(random.randint(1, 10))
print('A列表为：', A,
      'B列表为：', B)
"""
将A、B中的某两个元素相加
若二者之和为10
则输出该元素及该元素在列表中的对应位置
"""
for i in range(10):
    for j in range(10):
        if A[i]+B[j] == 10:
            print('A、B中的元素之和为10时，',
                  'A的位置为：', i,  
                  'B的位置为：', j, 
                  'A中对应元素的值为：', A[i],
                  'B中对应元素的值为：', B[j])
        else:
            print('A中第', i, '个位置整数不能与B中第', j, '个位置整数之和为10')
```

```python
# 随机生成两组线性可分的数据

# 导入随机数函数以供后面生成随机数使用
import random

# 建立x0,x1,x2,目标函数y,x1和x2中使得输出的y值为+1和-1的空列表
x0 = []
x1 = []
x2 = []
y = []
x1_yzheng = []
x2_yzheng = []
x1_yfu = []
x2_yfu = []

# 初始化i、j值，以供后面while循环使用
i = j = 0
# 随机产生权值w0,w1,w2，并打印出来
w0 = random.random()
w1 = -1*random.random()
w2 = random.random()
print('所选权值为：', w0, w1, w2)
print('生成的x0,x1,x2坐标分别为：')

# 计算初始设定的线性可分的直线（即感知器拟合的目标直线）的斜率和纵截距，以便与感知器迭代结果相对比
k = -w1 / w2
d = -w0 / w2
print('初始设定/目标直线斜率k=', k)
print('初始设定/目标直线纵截距d=', d)

# 用while循环生成矩阵X=(x0,x1,x2),并得到相应的输出矩阵Y
while i < 20:

    # 用random函数随机生成浮点数的x1,x2,并用round函数将其保留到小数点后一位
    temp_x1 = random.uniform(0, 5)
    temp_x1 = round(temp_x1, 1)
    temp_x2 = random.uniform(0, 5)
    temp_x2 = round(temp_x2, 1)

    # 计算0=w1*x1+w2*x2+w0的值
    y_output = w1 * temp_x1 + w2 * temp_x2 + w0

    # 分三种情况将相应的x1,x2,y按顺序依次存入列表
    if y_output > 0:
        x1.append(temp_x1)
        x2.append(temp_x2)
        y.append(1)
    elif y_output < 0:
        x1.append(temp_x1)
        x2.append(temp_x2)
        y.append(-1)
    else:
        x1.append(temp_x1)
        x2.append(temp_x2)
        y.append(0)
    x0.append(1)

    # 打印矩阵X=（x0,x1,x2），便于保存数据
    print(x0[i],  x1[i], x2[i])
    i += 1

# 用while循环将使y值分别为+1和-1的相应x1,x2存入四个列表，x1正、x2正,x1负、x2负以便于后面分颜色分别作图使用
while j < 20:
    if y[j] == 1:
        x1_yzheng.append(x1[j])
        x2_yzheng.append(x2[j])
    elif y[j] == -1:
        x1_yfu.append(x1[j])
        x2_yfu.append(x2[j])
    j += 1

# 打印生成的数据，便于保存
print('生成的使y为+1的x1的值为：', x1_yzheng)
print('生成的使y为+1的x2的值为：', x2_yzheng)
print('生成的使y为-1的x1的值为：', x1_yfu)
print('生成的使y为-1的x2的值为：', x2_yfu)
print('生成的y值：', y)
```

```python
# numpy 支持高级大量的维度数组与矩阵运算
import numpy as np
# Matplotlib 是一个 Python 的 2D绘图
import matplotlib.pyplot as plt

# 分别读取存在csv文件中的矩阵X=(x0,x1,x2),对应y值,x1正、x2正,x1负、x2负的数据
f1 = open("D:/BaiduNetdiskDownload/pythonProject1/Projiect1datas/f1x0+x1+x2--size20.csv", "rb")
temp_X = np.loadtxt(f1, delimiter=',', skiprows=0)
f1.close()
X = np.array(temp_X)

f2 = open("D:/BaiduNetdiskDownload/pythonProject1/Projiect1datas/f1y--size20.csv", "rb")
temp_Y = np.loadtxt(f2, delimiter=',', skiprows=0)
f2.close()
Y = np.array(temp_Y)

f3 = open("D:/BaiduNetdiskDownload/pythonProject1/Projiect1datas/f1x1_yzheng.csv", "rb")
x1_yzheng = np.loadtxt(f3, delimiter=',', skiprows=0)
f3.close()

f4 = open("D:/BaiduNetdiskDownload/pythonProject1/Projiect1datas/f1x2_yzheng.csv", "rb")
x2_yzheng = np.loadtxt(f4, delimiter=',', skiprows=0)
f4.close()

f5 = open("D:/BaiduNetdiskDownload/pythonProject1/Projiect1datas/f1x1_yfu.csv", "rb")
x1_yfu = np.loadtxt(f5, delimiter=',', skiprows=0)
f5.close()

f6 = open("D:/BaiduNetdiskDownload/pythonProject1/Projiect1datas/f1x2_yfu.csv", "rb")
x2_yfu = np.loadtxt(f6, delimiter=',', skiprows=0)
f6.close()

f7 = open("D:/BaiduNetdiskDownload/pythonProject1/Projiect1datas/f1w0+w1+w2.csv", "rb")
w_target = np.loadtxt(f7, delimiter=',', skiprows=0)
f7.close()

# 设定权值向量(w0,w1,w2),权值范围为-1,1
W = (np.random.random(3) - 0.5) * 2

# 设定学习率
lr = 0.01
# 初始化迭代次数，便于后面记录迭代次数
n = 0
# 神经网络输出
y_output = 0

# 定义update函数，w修正值
def update():
    global X, Y, W, lr, n
    n = n + 1
    y_output = np.sign(np.dot(X, W.T))
    # 计算权值差
    W_Tmp = lr * ((Y - y_output.T).dot(X))
    W = W + W_Tmp

# 用for循环进行感知器的迭代计算，寻找恰当的W
for index in range(10000):

    # 调用update函数
    update()
    # 每循环一次，打印每一次迭代的输出值
    y_output = np.sign(np.dot(X, W.T))

    # 用if进行判断，当假设函数与目标函数一致时，停止迭代，打印迭代次数，并跳出循环
    if (y_output == Y).all():
        print('Finished')
        print('epoch:', n)
        break

# 计算迭代结果的直线斜率与纵截距，并据此画出该假设函数直线
k = -W[1] / W[2]
d = -W[0] / W[2]
print('迭代所得权值w0,w1,w2为：', W)
print('迭代所得直线斜率k=', k)
print('迭代所得直线纵截距d=', d)
xdata1 = np.linspace(0, 5)
plt.figure()
plt.plot(xdata1, xdata1 * k + d, 'b', label='output_line')

# 计算目标结果的直线斜率与纵截距，并据此画出目标函数直线
k_target = -w_target[1] / w_target[2]
d_target = -w_target[0] / w_target[2]
print('目标权值w0,w1,w2为：', w_target)
print('目标直线斜率k=', k_target)
print('目标直线纵截距d=', d_target)
xdata2 = np.linspace(0, 5)
plt.plot(xdata2, xdata2 * k_target + d_target, 'r', label='target_line')
plt.legend()

# 描绘数据点
plt.scatter(x1_yzheng, x2_yzheng, c='red', s=2)
plt.scatter(x1_yfu, x2_yfu, c='green', s=2)

# 设置图表标题，并给坐标轴加上标签
plt.title('Procetron_size=1000_data4_lr=0.0001', fontsize=24)
plt.xlabel('x1_Value', fontsize=14)
plt.ylabel('x2_Value', fontsize=14)

plt.show()
    
```
