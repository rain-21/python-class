```python
#python基础练习
#字符串驻留机制，相同的字符串在满足以下条件时不会开辟新的空间创建，而是将已创建的地址再次赋值给变量
#但是在pycharm中已经对字符串进行了优化，强制让相同的字符串驻留
#条件如下（交互模式）：字符串长度为0或1；符合标识符的字符串；只在编译时进行驻留，而非运行时；[-5，256]中的整数
#第二条：只包括字母、数字、下划线的
#第三条解释如下
a='abc'
b='ab'+'c'
c=''.join(['ab','c'])#使用了join函数，正常来说是在运行时创建的，和a的地址不同，但是pycharm强制驻留了
print(a==c)#交互模式下应为False
#补充：字符串拼接时建议使用.join函数，效率比+高

#查询：index(),rindex(),find(),rfind()
a='hello world'
print(a.index(' '))
print(a.index('o'))#返回第一个子字符串的位置，位置从0开始
print(a.rindex('o'))#返回最后一个子字符串的位置
#print(a.index('k'))#会报错
print(a.find('k'))#不存在时返回-1，不会报错

#大小写转换，以下操作都会产生一个新的字符串，说明字符串不是可变序列
a='hello world,I love CN'
print(a.upper())#全变大写
print(a.lower())#全变小写
print(a.swapcase())#大小写互换
print(a.capitalize())#首字母大写，如果第一个不是字母则不变
print(a.title())#每一部分首字母大写

#字符串对齐
a='hello world'
print(a.center(20,'*'))#第一个参数为总宽度，小于字符串长度则返回字符串本身；第二个为填充标志，默认为空格
print(a.center(5))
print(a.ljust(20,'#'))
print(a.rjust(20,'*'))
print(a.zfill(20))#左侧填0，负数时在-号后添加
print('-666'.zfill(8))

#字符串劈分,split;rsplit,返回一个列表，标识为sep和maxsplit
a='hello world|learn python|i love cn'
print(a.split())#默认空格分隔
print(a.split(sep='|'))
print(a.split(sep='|',maxsplit=1))

#字符串的判断
print('abc张三Ⅳ'.isidentifier())#是否为合法标识符，字母数字下划线；汉字算字母，罗马数字算数字
print('b%'.isidentifier())
print('\t\n'.isspace())#是否为空白字符，包括回车，换行，水平制表符
print('abc张'.isalpha())#是否为字母
print('123'.isdecimal())#是否为十进制数字，罗马数字不算
print('12四百Ⅳ'.isnumeric())#是否为数字，汉字数字也可识别
print('abc12'.isalnum())#是否为字母和数字

#字符串替换与合并
a='hello python python python'
print(a.replace('python','java',2))#最后参数为最大替换次数，默认全换
b=['hello','world','python']
print(''.join(b))#b必须是列表或者元组，第一个为合并时的连接符
print('*'.join(b))
print('\t'.join(b))
print('*'.join('python'))

#字符串的比较
print(ord('A'))#求原始值，求ascii码的值，比较字符串大小时也是顺次比较这个值
print(ord('胡'))
print(chr(65))#根据ascii值求字符
print('apple'>'app')
#==比较value；is比较id地址

#字符串切片，和列表一样
a='hello world'
b=a[0::2]
print(b)

#格式化字符串
#%作为占位符，%s表示字符串；%i或者%d表示整数；%f表示浮点数
a='我叫%s,今年%d岁，存款%f元'%('胡家豪',20,10.5)#最后一个%为固定字符
print(a)
#输出结果如下：我叫胡家豪,今年20岁，存款10.500000元
#{}作为占位符
a='我叫{0},今年{1}岁，存款{2}元,我叫{0}'.format('胡家豪',20,10.5)#format中元素个数应该大于等于前面引用的
print(a)
#f-string
name='胡家豪'
age=20
a=f'我叫{name},今年{age}岁'
print(a)

#表示字符串精度和宽度
#%占位符
print('%10d'%99)#10表示宽度
print('hellohello')
print('%.3f'%3.1415926)#.3表示小数点后面的位数
print('%10.4f'%3.1415926)
#{}占位符
print('pi的值近似等于{0:10.5f}，而不是{1}'.format(3.1415926535,2.718281828))
print('pi的值近似等于{0:.5}，而不是{1}'.format(3.1415926535,2.718281828))#不带f表示总的位数

#字符串的编码转换
#编码：将字符串转化为二进制数据（bytes）
#解码：将二进制数据（bytes）转换为字符串类型
a='西安交通大学'
print(a.encode(encoding='GBK'))#GBK的编码格式表示一个汉字占两个字节
print(a.encode(encoding='utf-8'))#UTF-8的编码格式表示一个汉字占三个字节
print(b'\xce\xf7\xb0\xb2\xbd\xbb\xcd\xa8\xb4\xf3\xd1\xa7'.decode(encoding='GBK'))
print(b'\xe8\xa5\xbf\xe5\xae\x89\xe4\xba\xa4\xe9\x80\x9a\xe5\xa4\xa7\xe5\xad\xa6'.decode(encoding='utf-8'))
```

```python
#range函数
a=range(1,10)#不包括10
print(list(a))
a=range(1,10,2)
print(list(a))

#while循环
a=1
sum=0
while a<=100:
    sum+=a
    a+=1

print("和为",sum)

#for in 循环  ：for 自定义变量 in 可迭代对象
#可迭代对象有：字符串，一个整数序列
sum=0
for i in range(10):
    sum+=i
print(sum)


for i in "python":
    print(i)

#循环体中用不到自定义变量，则可用_代替
sum=0
for _ in range(10):
    print("python")


#练习：输出0-999中的水仙花数（个位，十位，百位的立方和等于这个数本身）
for i in range(100,1000):
    a=i//100
    b=(i-100*a)//10
    c=i-100*a-10*b
    if (a**3+b**3+c**3)==i:
        print(i)

#break语句，用于不正常的跳出循环，一般与if搭配使用

#break语句用于跳出之后的所有循环，continue语句用于跳出本次循环并返回循环入口进入下次循环

a=0
while a<3:
    s=input("输入一个值")
    if s=='123':
        print("输入正确")
        break
    else :
        a+=1
else:
    print("输入错误")

#嵌套循环打印乘法表
#知识点
print("a b")#空格
print("a    b")#tab,直接空4格
print("aaaabb")
print("a\tb")#从行首算起，每四个为一个单元，补齐当前单元
print("aa\tb")
print("aaaa\tb")
print("aaaaaaaab")
print("\na\nb",end="\t")
print("c")
print()

#程序如下
for i in range(1,10):
    for j in range(1,i+1):
        print(i,"*",j,"=",i*j,end='\t')
    print()

#多重循环中的break,continue语句只用于控制本层循环，而不对外层循环起作用

#continue与pass的区别：
for i in range(1,10):
    if i%2==0:
        pass
        print(i)

for i in range(1,10):
    if i%2==0:
        continue
        print(i)
```

```python
#运算符先后顺序：算术运算符>位运算>比较运算>bool运算>赋值运算符；有括号先算括号
#算数运算符中：幂运算>乘除运算>取商取余>加减

#**为幂运算
print(2**8)
a=10
b=20
a,b=b,a
print(a,b)

#=表示赋值运算，==表示值的比较运算，is表示id标识的比较运算，！=表示不等；is not同理
a,b=10,20
print(a==b)
a=10;b=10
print(a is b)
print(id(a),id(b))

[a,b]=[10,10]
print(a==b)
print(a is b)
print(id(a),id(b))

lst1=10
lst2=10
print(lst1 is lst2)

lst1=[10]
lst2=[10]
print(lst1 is lst2)
#对于10而言，加[]与不加[]是截然不同的

#特殊的bool运算符：in   not in,其他有and or not
s='helloworld'
print('e' in s)
print('k' not in s)

#位运算符    按位与&    按位或|   左移位<<  右移位>>
print(4&8)
print(4|8)
print(4<<1)
print(4>>1)
print(4<<7)
print(4>>3)
print(4>>4)
#4的二进制为00000100;8的二进制为00001000;经试验，左移4000位仍可以输出正常值，而不是0（思考原因）
```
