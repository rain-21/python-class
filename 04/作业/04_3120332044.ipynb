Face Recognition

```python
# face_recognition是一个强大、简单、易上手的人脸识别开源项目，并且配备了完整的开发文档和应用案例，特别是兼容树莓派系统。
# https://github.com/ageitgey/face_recognition
```

```python
# pip --default-timeout=1000 install face_recognition 先在命令行里pip也不行，失败了
import face_recognition
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np

image=face_recognition.load_image_file("akb48.jepg")
face_locations=face_recognition.face_locations(image)

im=np.array(Image.open('akb48.jepg'),dtype=np.uint8)

fit,ax=plt.subplots(1,figsize=(8,8))
ax.imshow(im)

for face_loc in face_locations:
    rect=patches.Rectangle(face_loc[1],face_loc[0],face_loc[3]-face_loc[1],face_loc[2]-face_loc[0],linewidth=2,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
plt.show()
```

SnowNLP: Simplified Chinese Text Processing
