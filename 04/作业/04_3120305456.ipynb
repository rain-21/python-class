```python
import re

line = "Cats are smarter than dogs"
matchObj = re.match( r'(.*) are (.*?) .*', line, re.M|re.I)
if matchObj:
   print ("matchObj.group() : ", matchObj.group())
   print ("matchObj.group(1) : ", matchObj.group(1))
   print ("matchObj.group(2) : ", matchObj.group(2))
else:
   print ("No match!!")
```

```python
import re
 
line = "Cats are smter than dogs"
 
searchObj = re.search( r'(.*) are (.*?) .*', line, re.M|re.I)
 
if searchObj:
   print ("searchObj.group() : ", searchObj.group())
   print ("searchObj.group(1) : ", searchObj.group(1))
   print ("searchObj.group(2) : ", searchObj.group(2))
else:
   print ("Nothing found!!")
```

```python
import re
 
line = "Cats are smarter than dogs"
 
matchObj = re.match( r'dogs', line, re.M|re.I)
if matchObj:
   print ("match --> matchObj.group() : ", matchObj.group())
else:
   print ("No match!!")
 
matchObj = re.search( r'dogs', line, re.M|re.I)
if matchObj:
   print ("search --> matchObj.group() : ", matchObj.group())
else:
   print ("No match!!")
```

```python
import requests
import re
from lxml import etree
from selenium import webdriver
from time import sleep
if __name__ == '__main__':
    url = 'https://www.youshuba.cc/ks/8/8337/'
    headers = {
        "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:83.0) Gecko/20100101 Firefox/83.0"
    }
    chapter_list_page = requests.get(url=url, headers=headers).text
    # print(chapter_list_page)
    ex = 'href="/ks/8/8337/(.*?).html"'
    chapter_url_list = re.findall(ex, chapter_list_page, re.S)
    for i in range(0, 3282):
        del chapter_url_list[0]
    # print(chapter_url_list)
    bro = webdriver.Firefox(executable_path="D:\IntelliJ IDEA 2018.2.5\Workspace\九龙至尊\geckodriver.exe")
    file = open('D:/九龙至尊.txt', 'w', encoding='utf-8')
    for chapter_url in chapter_url_list:
        chapter_url = 'https://www.youshuba.cc/ks/8/8337/'+ chapter_url +'.html'
        bro.get(chapter_url)
        sleep(2)
        chapter_page = bro.page_source
        tree = etree.HTML(chapter_page)
        # print(chapter_content)
        # print(tree)
        chapter_content = tree.xpath('//div[@id="chapter_content"]//text()')
        chapter_content = '\n'.join(chapter_content)
        # print(chapter_content)
        chapter_title = tree.xpath('//div[@id="chapter_title"]/h1/text()')[0]

        print(chapter_title + ' === 下载完成！！！')
#         file.write('\n' + chapter_title + '\n' + chapter_content + '\n')
    print("全书下载完成！！！")
    file.close()
    bro.close()
```

```python
import re
if __name__ == '__main__':
    url = "https://www.dingdiann.com/ddk26193/"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:83.0) Gecko/20100101 Firefox/83.0'
    }
    content_page = requests.get(url=url, headers=headers).text
    # print(content_page)
    ex = "/ddk26193/\d+.html"
    chapter_url_list = re.findall(ex, content_page, re.S)
    for i in range(14):
        del chapter_url_list[0]
    # print(chapter_url_list)
    fp = open("D:/九龙至尊.txt", 'w', encoding='utf-8')
    for chapter_url in chapter_url_list:
        chapter_url = "https://www.dingdiann.com" + chapter_url
        chapter_page = requests.get(url=chapter_url, headers=headers).text
        tree = etree.HTML(chapter_page)
        chapter_title = tree.xpath('//div[@class="bookname"]/h1/text()')[0]
        chapter_text = tree.xpath('//div[@id="content"]//text()')
        chapter_text = '\n'.join(chapter_text)
#         fp.write('\n' + chapter_title + '\n' + chapter_text + '\n')
        print(chapter_title + "=====下载完成！！！=====")

    print('九龙至尊下载完成！')
    fp.close()
```
