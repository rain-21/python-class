```python
class Animal:   #  父类
    def saywho(self):
        print('我是一个动物')
    pass
class Duck(Animal):    #子类
    def saywho(self,name):   # 重写父类的方法
        self.name=name
        print('我是一只%s'%self.name)
    pass
class Cat(Animal):    #子类
    def saywho(self,name):   # 重写父类的方法
        self.name=name
        print('我不是一只%s'%self.name)
    pass
def commenInvoke(obj):  # 统一调用的方法
    obj.saywho('鸭子')

duck1=Duck()
duck1.saywho('鸭子')
cat1=Cat()
cat1.saywho('小猫')
listobj=[Duck(),Cat()]
for item in listobj:
    commenInvoke(item)  # 循环调用函数
```

```python
class Animal:
    def __init__(self,name):
        self.name=name
        print('构造初始化方法')
        pass
    def __del__(self):
        print('析构方法')
        pass
    pass  # 自动调用，清理对象，内存空间被释放
cat = Animal('小猫')
del cat
#python 三大特征 封装 继承 多态 封装是把内容封装到某个地方，方便之后的使用，从另外一处调用
#继承  将多个类共有的方法提取到父类中，子类可以继承父类中公共属性和方法，提高效率
class Animal:
    def eat(self):
        print('吃饭了')
        pass
    def drink(self):
        print('喝水了')
        pass
    pass
class Dog(Animal):      # 继承Animal这个父类
    def wwj(self):
        print('小狗汪汪叫')
    pass
class Cat(Animal):
    def mmj(self):
        print('小猫喵喵叫')
        pass
    pass
d1=Dog()
d1.eat()
c1=Cat()
c1.drink()
#多继承
class shenxian:
    def fly(self):
        print('神仙都会飞')
        pass
    pass
class Monkey:
    def chitao(self):
        print('猴子吃桃')
    pass
class Wukong(shenxian,Monkey): # 既是神仙又是猴子
    pass
swk=Wukong()
swk.fly()
#当多个父类中有相同的方法时，广度优先，现在一个级别里面的父类里去查找，继承具有传递性，方法相同时，子类会覆盖父类
#当父类的方法已经不满足子类的需要时，子类就可以进行重写或完善
class Dog:
    def __init__(self,name,color):
        self.name=name
        self.color=color
        pass
    def bark(self):
        print('汪汪叫')
    pass
class Keji(Dog):
    def __init__(self,name,color):    # 如果没有重写，会执行父类的init函数，缺少函数就会报错
        #调用父类的方法
        # Dog.__init__(self,name,color)   #手动调用
        super(Keji, self).__init__(name,color)  # 另一种方法，自动找到父类
        #扩展其他属性
        self.hegiht=99
        pass
    def bark(self):
        super(Keji, self).bark()  # 自动调用父类的方法
        print('大声叫')
    def __str__(self):
        return '{}的颜色是{}'.format(self.name,self.color)
    pass
k1=Keji('柯基','红色')
k1.bark()
print(k1)
```
