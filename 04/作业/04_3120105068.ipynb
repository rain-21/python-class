```python
import random
N = random.randint(1,10)
teep = input("你能猜到小丸子心里在想什么数吗？那请输入一个1-10之间的整数：")
guess = int(teep)
if guess == N:
    print("哇塞！你猜对了！小丸子心里想的就是：")
else:
    while guess!=N:
        teep = input("不好意思，你猜错了哦！请重新输入：")
        guess = int(teep)
        if guess == N:
            print("哇塞！你猜对了！小丸子心里想的就是：")
        else:
            if guess > N:
                print("大了大了")
            else:
                print("小了小了")
print(N)
print("游戏结束，回家找妈妈")
```

```python
# 普通列表和列表解析
# 普通列表
# 创建空列表cubics
cubics = []
# 用for循环给列表cubics的后面每一次添加一个数的立方
for number in range(1,10):
    cubics.append(number**3)
print(cubics)
print('以下程序效果相同')

# 解析列表，很简洁，与上面程序的效果相同
cubics = [number**3 for number in range(1,10)]
print(cubics)
print(cubics[3:])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://archive.ics.uci.edu/ml/'
        'machine-learning-databases/iris/iris.data', header=None)
print(df.tail())
# extract sepal length and petal length
X = df.iloc[0:100, [0, 2]].values
# plot data
plt.scatter(X[:50, 0], X[:50, 1],
            color='red', marker='o', label='setosa')
plt.scatter(X[50:100, 0], X[50:100, 1],
            color='blue', marker='x', label='versicolor')
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
# plt.savefig('images/02_06.png', dpi=300)
plt.show()
```

```python
list = ['str_word', 'sentence is ok', 90, [9, 3, 2]]
def flat(l):
    f = []
    for x in l:
        if type(x) is list:
            f.extend(x)
        else:
            f.append(x)
    return f
print(flat(list))
```
