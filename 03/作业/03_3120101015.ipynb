```python
# 迭代器与生成器（iter和yeild）
turple = (1,2,3,4,5)
it = iter(turple)
# print(next(it))
# print(next(it))
# 用for循环遍历
for i in it:
    print(i,end = ' ')


dict = [1,2,3,4]
it = iter(dict)
print(next(it))

# yield生成器会暂停并保存当前所有的运行信息
import sys
def fibonacci(n):
    a,b,counter = 0,1,0
    while True:
        if counter>n:
            return
        yield a
        a,b = b,a+b
        counter += 1
f = fibonacci(10)

while True:
    try:
        print (next(f), end=" ")
    except StopIteration:
        sys.exit()
 

```

```python
# re.search
import re

pattern = re.compile(r'\d+')
m1 = pattern.search('one123')
print(m1)
m2 = pattern.search('one123',0,4)
print(m2)
print(m2.group())

m3 = re.search(r'\d+','one123')
print(m3)
print(m3.group())

m4 = re.search(r'[a-z]+','123Abcde',re.I)
print(m4)
print(m4.group())

# re.findall
pattern = re.compile(r'\d{2}')
m1 = pattern.findall('one1234')
print(m1)

m2 = pattern.findall('one123',0,4)
print(m2)

m3 = re.findall(r'\d+','one123')
print(m3)

m4 = re.findall(r'[a-z]+','123Abcde',re.I)
print(m4)

# re.split
pattern = re.compile(r'[\s\,\;]+')
m1 = pattern.split('a,b;; c    d')
print(m1)

m2 = re.split(r'[\s\,\;]+','a,b;; c    d')
print(m2)

# re.sub
s = 'hello 123 world 456'
pattern = re.compile(r'(\w+) (\w+)')
m1 = pattern.sub('hello world',s)
print(m1)

m2 = pattern.sub('hello world',s,1)
print(m2)

m3 = re.sub(r'(\w+) (\w+)','hello world',s,1)
m4 = re.sub(r'(\w+) (\w+)','hello world',s)
print(m3)
print(m4)
```

```python
# 正则表达式练习
#  E-mail 地址
import re
pattern = re.compile(r'^\w+([-+.]\w+)*@\w+([-.]\w+)*\.\w+([-.]\w+)*$')

def is_mail(email):
    if pattern.match(email):
        print("这是一个有效的Email地址：",email)
    else:
        print("这是一个无效的Email地址：",email)
is_mail("01xian_+jiaoda@163.com")
is_mail("01xianjiaoda@")
is_mail("xian.jiaoda@baidu.com")
is_mail("www.baidu.com")


# 18位身份证号码
pattern = re.compile(r'^((\d{18})|([0-9x]{18})|([0-9X]{18}))$')
def is_idcard(idcard):
    if pattern.match(idcard):
        print("这是一个身份证号码：",idcard)
    else:
        print("这不是一个身份证号码：",idcard)

is_idcard("411624200502091815")
is_idcard("41162420050-181815")
is_idcard("411624nfb8155566xx")
```

```python
# 面向对象
class Dog:
    def __init__(self,name):
        self.__name = name
        self.__age = None
        print(self.__name,"生成成功")
    
    def set_age(self,age):
        if not isinstance(age,int):
            print("输入的年龄必须是数字!")
            return False
        if age <=0:
            print("输入的年龄必须大于0!")
            return False
        self.__age = age
       

    def play(self):
        print("汪汪汪,我今年",self.__age,"岁")

dog = Dog("灰波")
dog.set_age("hello")
dog.set_age(-5)
dog.set_age(4)
dog.play()

# 类的继承
class Animal:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def play(self):
        print("我是%s"%self.name)
        print("我今年%d岁"%self.age)

class Dog(Animal):
    def __init__(self):
        super(Dog,self).__init__("机器狗",3)
class Cat(Animal):
    pass
class Pig(Animal):
    pass
dog = Dog()
cat =Cat("加菲猫",2)
pig =Pig("小猪佩奇",1)
dog.play()
cat.play()
pig.play()
```

```python
# 类方法
class complex:
    def __init__(self,real=0,image=0):
        self.real=real
        self.image=image

    @classmethod
    def add(cls,c1,c2):
        print(cls)
        c=complex()
        c.real=c1.real+c2.real
        c.image=c1.image+c2.image
        return c

if __name__=='__main__':
    c1=complex(1.2,2.3)
    c2=complex(1.6,2.4)
    c=complex.add(c1,c2)
    print("The final result is %.1f+%.1fi"%(c.real,c.image))


# super方法
class person:
    def __init__(self,name):
        print("person类构造方法被调用！")
        self.name=name
class student(person):
    def __init__(self,sno,name):
        print("student类构造方法被调用！")
        super().__init__(name)
        self.sno=sno
class postgraduate(student):
    def __init__(self,sno,name,tutor):
        print("postgraduate类构造方法被调用！")
        super().__init__(sno,name)
        self.tutor=tutor
if __name__=='__main__':
    pg=postgraduate("0225152645","小明","张三")
    print("该生的学号是%s,姓名是%s,指导教师是%s"%(pg.sno,pg.name,pg.tutor))
```
